```python
from nba_api.stats.endpoints import leaguedashplayerstats
import pandas as pd

# Fetch 2023-24 regular season stats
season = '2023-24'
player_stats = leaguedashplayerstats.LeagueDashPlayerStats(season=season, season_type_all_star='Regular Season')

# Convert to DataFrame
df = player_stats.get_data_frames()[0]

# Preview
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Drop irrelevant columns (you can add/remove as needed)
# Drop irrelevant or non-numeric columns
drop_cols = [
    'PLAYER_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'PLAYER_NAME',
    'NICKNAME', 'AGE', 'GP', 'W', 'L'
]

df_clean = df.drop(columns=drop_cols, errors="ignore")  # errors="ignore" = safe drop

# Keep player names for later display
player_names = df['PLAYER_NAME'].values

# Keep only numeric columns (in case others slipped through)
df_clean = df_clean.select_dtypes(include='number')

# Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(df_clean)

print("✅ Cleaned and scaled:", X.shape)


# Keep a copy of player names for reference
player_names = df['PLAYER_NAME'].values

# Standardize numeric stats
scaler = StandardScaler()
X = scaler.fit_transform(df_clean)

# Confirm shape: rows = players, columns = normalized stats
print("Shape of feature matrix:", X.shape)
```

```python
import umap
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Reduce to 2D with UMAP
reducer = umap.UMAP(random_state=42)
X_umap = reducer.fit_transform(X)

# Step 2: Cluster using KMeans
n_clusters = 8  # You can tweak this
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(X_umap)

# Step 3: Add to original DataFrame
df['RoleCluster'] = labels
df['x'] = X_umap[:, 0]
df['y'] = X_umap[:, 1]

# Step 4: Visualize
plt.figure(figsize=(10, 6))
for i in range(n_clusters):
    cluster = df[df['RoleCluster'] == i]
    plt.scatter(cluster['x'], cluster['y'], label=f"Role {i}", alpha=0.6)

plt.title("NBA Player Archetypes (UMAP + KMeans)")
plt.xlabel("UMAP Dimension 1")
plt.ylabel("UMAP Dimension 2")
plt.legend()
plt.show()
```

```python
# Group player roles by team
team_roles = df.groupby(['TEAM_ABBREVIATION', 'RoleCluster']).size().unstack(fill_value=0)

# Normalize to get role distribution % per team
team_role_dist = team_roles.div(team_roles.sum(axis=1), axis=0)

# Show role breakdown for a specific team
def show_team_roles(team_abbr):
    print(f"🧩 Role distribution for {team_abbr}:")
    display(team_role_dist.loc[team_abbr].sort_values(ascending=False))

# Example usage:
show_team_roles("BOS")  # Replace with any team code: LAL, DEN, MIA, etc.
```

```python
def recommend_trade_targets(team_abbr, top_n=3):
    # 1. Get team role distribution
    team_dist = team_role_dist.loc[team_abbr]

    # 2. Compare to league average to find weakest roles
    league_avg = team_role_dist.mean()
    missing_roles = (league_avg - team_dist).sort_values(ascending=False)

    print(f"🔍 {team_abbr} is underrepresented in these roles:")
    display(missing_roles.head(top_n))

    # 3. Recommend players in that role, from other teams
    for role in missing_roles.head(top_n).index:
        print(f"\n🎯 Suggested players for missing Role {role}:")
        candidates = df[(df['RoleCluster'] == role) & (df['TEAM_ABBREVIATION'] != team_abbr)]
        display(candidates[['PLAYER_NAME', 'TEAM_ABBREVIATION', 'RoleCluster']].head(5))

# Example usage:
recommend_trade_targets("BOS")  # Try with any team abbreviation
```

```python
def audit_team(df, team_abbr):
    team_df = df[df['TEAM_ABBREVIATION'] == team_abbr].copy()

    report = {}

    # Rule 1: Two 7-footers
    report["Two 7-footers"] = (team_df['HEIGHT'] >= 84).sum() >= 2

    # Rule 2: Frontcourt avg height
    frontcourt = team_df[team_df['POSITION'].isin(['C', 'PF', 'F', 'FC'])]
    report["Frontcourt avg height > 6'5"] = frontcourt['HEIGHT'].mean() >= 77

    # Rule 3: Young core
    report["7+ players under 27"] = (team_df['AGE'] < 27).sum() >= 7

    # Rule 4: Role balance
    defense_clusters = [0, 3, 5]
    offense_clusters = [1, 2, 4, 6, 7]
    report["3+ defenders"] = (team_df['RoleCluster'].isin(defense_clusters)).sum() >= 3
    report["5+ offensive players"] = (team_df['RoleCluster'].isin(offense_clusters)).sum() >= 5

    # Rule 5: Confident vet shooter
    vet = team_df[team_df['AGE'] >= 30]
    report["Vet: FT% > 82 & 3P% > 36"] = any((vet['FT_PCT'] > 0.82) & (vet['FG3_PCT'] > 0.36))

    # Rule 6: Midrange masters
    team_df['FG2_PCT'] = team_df['FG_PCT'] - team_df['FG3_PCT']
    report["2+ players FG2% > 55"] = (team_df['FG2_PCT'] > 0.55).sum() >= 2

    return report
```

```python
audit = audit_team(df, "BOS")
for k, v in audit.items():
    print(f"{'✅' if v else '❌'} {k}")
```

```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import commonplayerinfo
import time

def get_player_height(player_name):
    try:
        search = players.find_players_by_full_name(player_name)
        if not search:
            return None
        player_id = search[0]['id']
        data = commonplayerinfo.CommonPlayerInfo(player_id=player_id).get_normalized_dict()
        height_str = data['CommonPlayerInfo'][0]['HEIGHT']  # Format: "6-10"
        feet, inches = map(int, height_str.split('-'))
        return feet * 12 + inches
    except:
        return None

# Run once, takes a few minutes
df['HEIGHT'] = df['PLAYER_NAME'].apply(lambda x: get_player_height(x))
```

```python
def evaluate_team(df, team_abbr):
    team = df[df['TEAM_ABBREVIATION'] == team_abbr]

    report = []

    # 1. At least two 7-footers (height >= 84 inches)
    seven_footers = team[team['HEIGHT'] >= 84]
    report.append(f"{'✅' if len(seven_footers) >= 2 else '❌'} 7-footers: {len(seven_footers)} found")

    # 2. Frontcourt average height above 6'5" (77 inches)
    frontcourt_keywords = ['F', 'C', 'FC', 'PF', 'SF', 'C-F', 'F-C']
    frontcourt = team[team['POSITION'].astype(str).str.upper().str.contains('|'.join(frontcourt_keywords))]
    avg_height = frontcourt['HEIGHT'].mean()
    report.append(f"{'✅' if avg_height >= 77 else '❌'} Avg frontcourt height: {avg_height:.1f} in")

    # 3. At least 7 players under 27
    young_players = team[team['AGE'] < 27]
    report.append(f"{'✅' if len(young_players) >= 7 else '❌'} Young players (<27): {len(young_players)}")

    # 4. At least 3 defensive specialists
    defense_stats = ['STL', 'BLK', 'DEF_RATING', 'REB']
    team['DEF_SCORE'] = team[['STL', 'BLK']].sum(axis=1)  # extend if you have more stats
    def_specialists = team[team['DEF_SCORE'] > team['DEF_SCORE'].quantile(0.7)]
    report.append(f"{'✅' if len(def_specialists) >= 3 else '❌'} Defensive players: {len(def_specialists)}")

    # 5. At least 5 offensive-oriented players
    team['OFF_SCORE'] = team[['PTS', 'AST', 'FG_PCT']].sum(axis=1)  # optional: include 3PT_PCT
    off_specialists = team[team['OFF_SCORE'] > team['OFF_SCORE'].quantile(0.7)]
    report.append(f"{'✅' if len(off_specialists) >= 5 else '❌'} Offensive players: {len(off_specialists)}")

    # Bonus: Show key metrics
    report.append("\nTop Offensive Player:")
    report.append(off_specialists[['PLAYER_NAME', 'PTS', 'AST']].sort_values(by='OFF_SCORE', ascending=False).head(1).to_string(index=False))

    report.append("\nTop Defensive Player:")
    report.append(def_specialists[['PLAYER_NAME', 'STL', 'BLK']].sort_values(by='DEF_SCORE', ascending=False).head(1).to_string(index=False))

    return "\n".join(report)
```

```python
def evaluate_team(df, team_abbr):
    report = []
    team = df[df['TEAM_ABBREVIATION'] == team_abbr]

    if team.empty:
        return f"Team {team_abbr} not found."

    # 🏀 Rule 1: At least 2 players who are 7 feet (84 inches) or taller
    tall_players = team[team['HEIGHT'] >= 84]
    report.append(f"{'✅' if len(tall_players) >= 2 else '❌'} 7-footers: {len(tall_players)}")

    # 🧱 Rule 2: Avg frontcourt height > 6'5" (77 inches)
    frontcourt = team[team['HEIGHT'] >= 77]
    avg_height = frontcourt['HEIGHT'].mean()
    report.append(f"{'✅' if avg_height >= 77 else '❌'} Avg frontcourt height: {avg_height:.1f} in")

    # 🛡️ Rule 3: At least 3 defense-oriented players
    defense_stats = team[['STL', 'BLK', 'DREB']].mean(axis=1)
    defense_top3 = team.loc[defense_stats.nlargest(3).index]
    report.append(f"✅ Defensive anchors: {[name for name in defense_top3['PLAYER_NAME']]}")

    # 🔥 Rule 4: At least 5 offense-oriented players
    offense_stats = team[['PTS', 'AST', 'FG_PCT']].mean(axis=1)
    offense_top5 = team.loc[offense_stats.nlargest(5).index]
    report.append(f"✅ Offensive weapons: {[name for name in offense_top5['PLAYER_NAME']]}")

    # 👶 Rule 5: At least 7 players under 27
    young_players = team[team['AGE'] < 27]
    report.append(f"{'✅' if len(young_players) >= 7 else '❌'} Young players (<27): {len(young_players)}")

    return "\n".join(report)
```

```python
def suggest_trade_targets(df, team_abbr):
    team = df[df['TEAM_ABBREVIATION'] == team_abbr]
    other_teams = df[df['TEAM_ABBREVIATION'] != team_abbr]

    suggestions = []

    # Need 7-footers
    if (team['HEIGHT'] >= 84).sum() < 2:
        candidates = other_teams[other_teams['HEIGHT'] >= 84]
        top = candidates.sort_values(by='BLK', ascending=False).head(3)
        suggestions.append("📏 Need 7-footers:\n" + "\n".join(top['PLAYER_NAME'].tolist()))

    # Avg frontcourt height < 77"
    frontcourt = team[team['HEIGHT'] >= 77]
    if frontcourt['HEIGHT'].mean() < 77:
        candidates = other_teams[other_teams['HEIGHT'] >= 77]
        top = candidates.sort_values(by='DREB', ascending=False).head(3)
        suggestions.append("🏗️ Frontcourt height too low:\n" + "\n".join(top['PLAYER_NAME'].tolist()))

    # Too few defenders
    defense_stats = team[['STL', 'BLK', 'DREB']].mean(axis=1)
    if len(defense_stats[defense_stats > 1.5]) < 3:
        candidates = other_teams.copy()
        candidates['DEFENSE_SCORE'] = candidates[['STL', 'BLK', 'DREB']].mean(axis=1)
        top = candidates.sort_values(by='DEFENSE_SCORE', ascending=False).head(3)
        suggestions.append("🛡️ Need more defenders:\n" + "\n".join(top['PLAYER_NAME'].tolist()))

    # Too few offensive weapons
    offense_stats = team[['PTS', 'AST', 'FG_PCT']].mean(axis=1)
    if len(offense_stats[offense_stats > 15]) < 5:
        candidates = other_teams.copy()
        candidates['OFFENSE_SCORE'] = candidates[['PTS', 'AST', 'FG_PCT']].mean(axis=1)
        top = candidates.sort_values(by='OFFENSE_SCORE', ascending=False).head(3)
        suggestions.append("🔥 Need more scorers:\n" + "\n".join(top['PLAYER_NAME'].tolist()))

    # Not enough young players
    if (team['AGE'] < 27).sum() < 7:
        candidates = other_teams[other_teams['AGE'] < 27]
        top = candidates.sort_values(by='PTS', ascending=False).head(3)
        suggestions.append("👶 Need more young talent:\n" + "\n".join(top['PLAYER_NAME'].tolist()))

    return "\n\n".join(suggestions) if suggestions else "✅ No obvious weaknesses. Killer squad."
```

```python
print(suggest_trade_targets(df, "GSW"))
```
