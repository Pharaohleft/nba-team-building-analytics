```python
from pytrends.request import TrendReq
import pandas as pd
import time

# Load your trends file
df = pd.read_csv("consumer_trend_stock_correlations.csv")

# Set up pytrends
pytrends = TrendReq(hl='en-US', tz=360)

# Store results
all_trends = {}

# Loop through each trend keyword
for trend in df['trend']:
    try:
        pytrends.build_payload([trend], timeframe='today 12-m')
        data = pytrends.interest_over_time()

        if not data.empty:
            all_trends[trend] = data[trend]
            print(f"‚úÖ Pulled: {trend}")
        else:
            print(f"‚ö†Ô∏è No data for: {trend}")

        time.sleep(1)  # Google rate limit handling
    except Exception as e:
        print(f"‚ùå Error with {trend}: {e}")

# Combine all trends into one DataFrame
trend_df = pd.DataFrame(all_trends)
trend_df.to_csv("trend_interest_timeseries.csv")

print("‚úÖ Saved trend_interest_timeseries.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
import yfinance as yf

# STEP 1: Fetch Google Trends data
pytrends = TrendReq()
pytrends.build_payload(kw_list=["ChatGPT"], timeframe="today 12-m")
trend_df = pytrends.interest_over_time()

# Extract trend
trend_series = trend_df["ChatGPT"]

# STEP 2: Fetch MSFT stock price and resample to weekly
stock_df = yf.download("MSFT", start="2023-07-01", end="2024-07-26", auto_adjust=False)

stock_series = stock_df["Adj Close"].resample("W").mean()

# STEP 3: Align both datasets on date
aligned_df = pd.concat([trend_series, stock_series], axis=1, join="inner").dropna()
trend_aligned = aligned_df.iloc[:, 0]
stock_aligned = aligned_df.iloc[:, 1]

# STEP 4: Lagged correlation function
def lagged_correlation(trend_series, stock_series, max_lag=15):
    results = []
    for lag in range(-max_lag, max_lag + 1):
        shifted = trend_series.shift(lag)
        corr = shifted.corr(stock_series)
        results.append((lag, corr))
    return pd.DataFrame(results, columns=["lag_days", "correlation"])

# STEP 5: Run lagged correlation
corr_df = lagged_correlation(trend_aligned, stock_aligned)

# STEP 6: Plot the correlation
plt.figure(figsize=(10, 5))
plt.plot(corr_df["lag_days"], corr_df["correlation"], marker='o', label='correlation')
plt.axvline(0, color='gray', linestyle='--')
plt.title("Lagged Correlation: ChatGPT vs MSFT")
plt.xlabel("Lag (days)")
plt.ylabel("Correlation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a fake trend with a rising spike
np.random.seed(0)
dates = pd.date_range("2024-01-01", "2024-06-30", freq="W")
trend = np.random.normal(loc=50, scale=10, size=len(dates))
trend[10:15] += 20  # spike

# Create stock data that lags the trend by 3 weeks
stock = np.roll(trend, 3) + np.random.normal(0, 2, len(dates))  # add small noise

# Create series
trend_series = pd.Series(trend, index=dates, name="Synthetic Trend")
stock_series = pd.Series(stock, index=dates, name="Synthetic Stock")

# Lagged correlation function
def lagged_correlation(trend_series, stock_series, max_lag=15):
    results = []
    for lag in range(-max_lag, max_lag + 1):
        shifted = trend_series.shift(lag)
        corr = shifted.corr(stock_series)
        results.append((lag, corr))
    return pd.DataFrame(results, columns=["lag_days", "correlation"])

# Run it
corr_df = lagged_correlation(trend_series, stock_series)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(corr_df["lag_days"], corr_df["correlation"], marker='o')
plt.axvline(0, color='gray', linestyle='--')
plt.title("Synthetic Trend vs Lagged Stock (Should Peak at +3)")
plt.xlabel("Lag (weeks)")
plt.ylabel("Correlation")
plt.grid(True)
plt.show()
```
